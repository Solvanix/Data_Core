```python
import os

# إعدادات المسار والجذر
ROOT_DIR = r"C:\Users\ali\Desktop\Solvanix_Local_Hub\Data_Core"
EXCLUDE_PATTERNS = [r"page_\d+", r"image_\d+", r"\.jpg$", r"\.png$", r"\.txt$"]

import re

def should_exclude(filename):
    return any(re.search(pattern, filename) for pattern in EXCLUDE_PATTERNS)

def print_tree(startpath, prefix=""):
    for item in sorted(os.listdir(startpath)):
        full_path = os.path.join(startpath, item)
        if should_exclude(item):
            continue
        if os.path.isdir(full_path):
            print(f"{prefix}📁 {item}/")
            print_tree(full_path, prefix + "│   ")
        else:
            print(f"{prefix}📄 {item}")

# تشغيل السكربت
print(f"📦 شجرة المستودع: {ROOT_DIR}\n")
print_tree(ROOT_DIR)
```

```python
import os
import re

# المسار إلى الملف النصي
input_path = r"C:\Users\ali\Desktop\Solvanix_Local_Hub\Data_Core\Processed_Sources\nird_islamic_culture_curriculum_comprehensive_alasriyya_extracted.txt"

# استخراج اسم الملف بدون الامتداد
filename = os.path.splitext(os.path.basename(input_path))[0]
output_path = fr"C:\Users\ali\Desktop\Solvanix_Local_Hub\Data_Core\Interfaces\{filename}.html"

# قراءة الملف
with open(input_path, "r", encoding="utf-8") as file:
    content = file.read()

# تقسيم المحتوى إلى صفحات
pages = re.split(r"--- الصفحة (\d+) ---", content)

# بناء HTML
html_parts = [
    "<!DOCTYPE html>",
    "<html lang='ar'>",
    "<head>",
    "<meta charset='UTF-8'>",
    f"<title>{filename}</title>",
    "<style>",
    "body { font-family: 'Segoe UI', sans-serif; direction: rtl; background: #f0f0f0; padding: 20px; }",
    "details { margin-bottom: 12px; border: 1px solid #ccc; padding: 10px; border-radius: 6px; transition: background 0.3s ease; }",
    "summary { font-weight: bold; cursor: pointer; font-size: 1.1em; }",
    ".odd { background: #ffffff; }",
    ".even { background: #f7f7f7; }",
    "</style>",
    "</head>",
    "<body>",
    f"<h1>{filename}</h1>"
]

# الصفحات تبدأ من index 1 لأن index 0 هو المحتوى قبل أول صفحة (غالبًا فارغ)
for i in range(1, len(pages), 2):
    page_number = int(pages[i].strip())
    page_content = pages[i + 1].strip().replace("\n", "<br>")
    css_class = "odd" if page_number % 2 != 0 else "even"
    html_parts.append(f"<details class='{css_class}'><summary>الصفحة {page_number}</summary><div>{page_content}</div></details>")

html_parts.append("</body></html>")

# حفظ الملف
with open(output_path, "w", encoding="utf-8") as file:
    file.write("\n".join(html_parts))

print(f"✅ تم إنشاء الملف بنجاح: {output_path}")
```

```python
import os
import shutil
import re
import json

# إعدادات المسارات
SOURCE_DIR = "/mnt/data"
ARCHIVE_DIR = os.path.join(SOURCE_DIR, "symbolic_page_archive")

# إنشاء مجلد الأرشيف إن لم يكن موجودًا
os.makedirs(ARCHIVE_DIR, exist_ok=True)

# نمط التعرف على معرف الصفحة من اسم الملف
page_pattern = re.compile(r"(page[_]?\d+)", re.IGNORECASE)

# خريطة لتجميع الملفات حسب الصفحة
page_files = {}

# المرور على كل الملفات في المصدر
for filename in os.listdir(SOURCE_DIR):
    filepath = os.path.join(SOURCE_DIR, filename)
    if os.path.isfile(filepath):
        match = page_pattern.search(filename)
        if match:
            page_id = match.group(1).lower()
            page_files.setdefault(page_id, []).append(filename)

# نقل الملفات إلى مجلدات فرعية
for page_id, files in page_files.items():
    page_dir = os.path.join(ARCHIVE_DIR, page_id)
    os.makedirs(page_dir, exist_ok=True)
    for fname in files:
        src = os.path.join(SOURCE_DIR, fname)
        dst = os.path.join(page_dir, fname)
        shutil.copy2(src, dst)

# حفظ ملخص الربط
summary_path = os.path.join(ARCHIVE_DIR, "summary.json")
with open(summary_path, "w", encoding="utf-8") as f:
    json.dump(page_files, f, indent=2, ensure_ascii=False)

print(f"✅ تم تنظيم الملفات في: {ARCHIVE_DIR}")
print(f"📄 تم حفظ الملخص في: {summary_path}")
```

```python
def validate_symbolic_links(chapter_map, page_map, text_map):
    report = {
        "صفحات غير موجودة في page_map": [],
        "وحدات غير موجودة في text_map": [],
        "صفحات بدون وحدات": [],
        "وحدات بدون نص": []
    }

    for chapter_id, chapter_data in chapter_map.items():
        pages = chapter_data.get("pages", [])
        for page_id in pages:
            page_id_str = str(page_id)
            if page_id_str not in page_map:
                report["صفحات غير موجودة في page_map"].append(page_id)
                continue

            units = page_map.get(page_id_str, [])
            if not units:
                report["صفحات بدون وحدات"].append(page_id)

            for unit_id in units:
                if unit_id not in text_map:
                    report["وحدات غير موجودة في text_map"].append(unit_id)
                elif not text_map[unit_id].strip():
                    report["وحدات بدون نص"].append(unit_id)

    return report
```

```python
def clean_label(label):
    corrections = {
        "الثائي": "الثاني",
        "الأؤل": "الأول",
        "الرابج": "الرابع",
        "الراببع": "الرابع",
        "السابح": "السابع",
        "الشاني": "الثاني",
        "الثائث": "الثالث",
        "الخالث": "الثالث",
        "المراببع": "الرابع",
        "الشالث": "الثالث",
        "الرابيع": "الرابع",
        "القانفي": "الثاني",
        "سنتحدث": "غير معرف",
        "المهم": "غير معرف"
    }
    for wrong, right in corrections.items():
        label = label.replace(wrong, right)
    return label.strip()
```

```python
from fuzzywuzzy import process

known_labels = [
    "المبحث الأول", "المبحث الثاني", "المبحث الثالث", "المبحث الرابع", "المبحث الخامس",
    "المطلب الأول", "المطلب الثاني", "المطلب الثالث", "المطلب الرابع", "الفصل الأول",
    "الفصل الثاني", "الفصل الثالث", "الفصل الرابع", "الفصل الخامس"
]

def clean_label_fuzzy(label):
    label = label.strip()
    best_match, score = process.extractOne(label, known_labels)
    return best_match if score > 80 else "غير معرف"
```
